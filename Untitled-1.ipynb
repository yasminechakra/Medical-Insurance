#### Data Processing 

```python
#encoding categorical variables (sex, smoker, region) into numerical format
#encoding the sex column bu assigning numerical values to the categories male and female

#label encode the "sex" column (Label Encoding)
df["sex"] = df["sex"].map({"male" : 0, "female" : 1})

#verify the encoding 
print(df["sex"].head())
```

```python
#encoding the smoker column
df["smoker"] = df["smoker"].map({"yes" : 1, "no" : 0})

#verify the encoding
print(df["smoker"].head())
```

```python
#One-Hot Encoding for the "region" column
df = pd.get_dummies(df, columns = ["region"], prefix= "region")

print(df.head())
```

```python
df.head()
```

```python
#Scaling continuous features like age, bmi, and charges
#applying standardization 

#initialize thz scaler
scaler = MinMaxScaler()

#aapply scaling tp continous features
df[["age", "bmi", "charges"]] = scaler.fit_transform(df[["age", "bmi", "charges"]])

#check the transformed data 
print(df[["age", "bmi", "charges"]].head())
```

```python
#Split the data into training and testing sets 
x = df.drop(columns = ["charges"])
y = df["charges"]

#split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

#check the size of the resulting datasets
print(f"Training data : {x_train.shape}, {y_train.shape}")
print(f"Testing data: {x_test.shape}, {y_test.shape}")
```
